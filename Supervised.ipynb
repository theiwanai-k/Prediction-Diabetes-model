## Point-Biserial Correlation Coefficient (PBCC) 

is a statistical method to measure the strength and direction of the relationship between:

One continuous (numeric) variable

One binary (dichotomous) variable

If r = 0.65, p < 0.05 → strong positive relation → customers who churn (1) tend to have higher monthly charges.

If r = -0.4, p < 0.05 → moderate negative relation → customers who churn have lower monthly charges.

If p > 0.05 → relationship is not statistically significant.

### Outliers

#### Remove

If outliers are clearly invalid data points.

Example: Age = -10, BMI = 300.

#### Cap / Flooring (Winsorization)

Replace extreme values with reasonable boundaries (e.g., 1st and 99th percentile).

Example: If Salary > 1M, set it = 1M.

#### Transformation

Use log, square root, Box-Cox transforms to reduce skewness.

Example: Income data is highly skewed → apply log transformation.

#### Imputation

Replace with mean/median if suspected as error.

Example: Replace negative house prices with median house price.

#### Separate Model Treatment

Train a different model for normal vs extreme cases.

Example: Build separate pricing model for regular houses vs luxury villas.

#### Use Robust Models

Instead of removing, use models less sensitive to outliers:

Decision Trees / Random Forests

XGBoost / LightGBM

Robust Regression (Huber, RANSAC).

```python
x_train, x_test, y_train, y_test = train_test_split(Q,R,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python
acc = accuracy_score(y_test, y_predict)
prec = precision_score(y_test, y_predict)
rec = recall_score(y_test, y_predict)
f1 = f1_score(y_test, y_predict)
roc = roc_auc_score(y_test, y_predict)

# 6. Print results
print("📊 Classification Evaluation Metrics")
print("----------------------------------")
print(f"Accuracy   : {acc:.4f}")
print(f"Precision  : {prec:.4f}")
print(f"Recall     : {rec:.4f}")
print(f"F1 Score   : {f1:.4f}")
print(f"ROC-AUC    : {roc:.4f}")

# 7. Extra: confusion matrix and detailed report
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_predict))

print("\nClassification Report:")
print(classification_report(y_test, y_predict))
```
