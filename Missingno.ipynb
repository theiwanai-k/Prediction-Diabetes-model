```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import missingno as msno
```

***About Dataset***
****Context***
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective is to predict based on diagnostic measurements whether a patient has diabetes.

***Content***
Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

Pregnancies: Number of times pregnant
Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
BloodPressure: Diastolic blood pressure (mm Hg)
SkinThickness: Triceps skin fold thickness (mm)
Insulin: 2-Hour serum insulin (mu U/ml)
BMI: Body mass index (weight in kg/(height in m)^2)
DiabetesPedigreeFunction: Diabetes pedigree function
Age: Age (years)
Outcome: Class variable (0 or 1)

### missingno
is a Python library that provides visualizations to understand missing data patterns.
Instead of just looking at df.isnull().sum(), it helps you see:

- Where missing values occur

- Which variables have more missing data

- Correlations in missingness between columns

```python
"""A correlation heatmap of missing values.

High correlation between two columns = they tend to be missing together.

Example: If Salary and Experience always miss together  strong correlation."""



msno.heatmap(df)
plt.show()
```

```python
data = {"Name":['a','b','c','d','d'],"Age":[25,None,22,None,40],"Salary":[50000,None,None,40000,None]}
data1 = pd.DataFrame(data)
data1
```

```python
msno.heatmap(data1)
plt.show()
```

```python
""" Matrix plot - Each column is a vertical bar.

White lines = missing values.

Helps you quickly see where missing values occur."""


msno.matrix(data1) 
plt.show()
```

```python
""" Bar Plot - Each bar = number of non-null values per column.

Gray portion = missing values count.

Easy to see which feature is most incomplete."""



msno.bar(data1)
plt.show()
```

```python
"""Groups features with similar missing data patterns.

Useful if your dataset is wide (many features)  you can cluster columns based on missingness."""


msno.dendrogram(data1)
plt.show()
```

matrix()  Quick overview (where missing data occurs in dataset).

bar()  See how much data is missing in each feature.

heatmap()  Detect relationships between missingness (helps in deciding imputation strategies).

dendrogram()  Find clusters of columns with similar missing patterns.
